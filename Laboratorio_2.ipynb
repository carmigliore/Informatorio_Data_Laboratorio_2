# **9. Rellenar registros de la columna clasificación CIE10**



---
**9.1 Verificar que codigos cie10 de la columna "cie10_causa_id" no tienen su nombre en la columna "cie10_clasificacion"**


```python
# Creo un dataframe de solo los registros nulos del campo 'cie10_clasificacion' pero con registros 'cie10_causa_id'
df_sin_columnas_provincia_sin_cie10 = df_sin_columnas_provincia[df_sin_columnas_provincia['cie10_clasificacion'].isnull() &
                                                                df_sin_columnas_provincia['cie10_causa_id'].notnull()]

# Selecciono solo la columna 'cie10_causa_id' y listo los registros que no tienen su correlato en 'cie10_clasificacion'
causas_sin_clasificacion = df_sin_columnas_provincia_sin_cie10[['cie10_causa_id']].groupby(['cie10_causa_id']).count().index

# Listo los codigos que no tienen su correlato descriptivo
display(causas_sin_clasificacion)
```

---
**9.2 Completar clasificacion cie10 según código cie10**

Se detectaron que algunos codigos cie10 identificados en 9.1 tienen su nombre en la columna cie10_clasificacion, pero en algunos registro faltan (missing value)

Por esto se completarán los registros de cie10_clasificacion segun el valor de cie10_causa_id


```python
# Crear un mapeo de cie10_causa_id a cie10_clasificacion
mapping = df_sin_columnas_provincia.dropna(subset=['cie10_clasificacion']).set_index('cie10_causa_id')['cie10_clasificacion'].to_dict()

# Rellenar los valores nulos en 'cie10_clasificacion' usando el mapeo
df_sin_columnas_provincia['cie10_clasificacion'] = df_sin_columnas_provincia['cie10_clasificacion'].fillna(df_sin_columnas_provincia['cie10_causa_id'].map(mapping))

# Verificar si aún quedan nulos en 'cie10_clasificacion'
print("La cantidad de registros nulos en la columna 'cie10_clasificacion' después de rellenar es de:", df_sin_columnas_provincia['cie10_clasificacion'].isnull().sum())
print(f"Lo que solo representa el {round((df_sin_columnas_provincia['cie10_clasificacion'].isnull().sum() / df_sin_columnas_provincia['cie10_clasificacion'].count() * 100),2)} % del total de registros del dataframe")
```

---
**9.3 Borrar registros cie10_clasificación nulos o vacios**

Se observa que todavia quedan 1611 registros sin datos en la columna cie10_clasificacion, por lo que se deduce no tienen su correlato descriptivo. Se podría recurrir al clasificador internacional y completar manualmente los datos pero: dado que son 5 códigos y que  representan solo el 0,2% de los registros se opta por borrarlos.

Con esto se termina de resolver la discrepancia 5.2.2

```python
# Borrar registros nulos de la columna
df_sin_columnas_provincia_cie10 = df_sin_columnas_provincia.dropna(subset=['cie10_clasificacion'])
x = df_sin_columnas_provincia_cie10['cie10_clasificacion'].isnull().sum()
print(f"La cantidad de registros nulos en la columna 'cie10_clasificacion' después de: {x}")
```

**7. Correlación de las 10 principales patología por grupo etareo**

Se observa:
1. La enfermedad "diabetes mellitus no especificada" tiene una alta correlación con el Grupo Etareo de 55 a 74 años, en segundo lugar el grupo mayor a 75 años y en tercer lugar de 35 a 54 años
2. El grupo etareo de 55 a 74 años tiene una mediana correlación con las causas de muerte: insuficiencia cardíaca, diabetes mellitus no insulinodependeiente y Aneurisma con disección aortica. Este grupo etáreo es el que tiene mayor correlación con las causas de muerte.
3. El grupo etareo de 15 a 34 años tiene una alta correlación con la causa de muerte: lesion autoinflingida intencionalmente por ahorcamiento y estrangulamiento.
4. Las enfermedades Neumonia y otros trastornos respiratorios tienen mediana correlación y en foma similar en todos los grupos etareos.

```python
# Crear un dataframe que excluye la edad sin especificar
edad_sin_especificar = '06.Sin especificar'
df_sin_especificar_edad = df_sexo_depurado[df_sexo_depurado['grupo_edad']!=edad_sin_especificar]

# Crear una tabla de contingencia de cie10_clasificacion y grupo_edad
tabla_contingencia = pd.crosstab(df_sin_especificar_edad['grupo_edad'], df_sin_especificar_edad['cie10_clasificacion'])

# Seleccionar las 10 principales clasificaciones CIE10 para el gráfico
top_10_cie10 = df_sin_especificar_edad['cie10_clasificacion'].value_counts().sort_values(ascending=False).head(10).index.tolist()
tabla_contingencia_top_10 = tabla_contingencia[top_10_cie10]

# Crear un mapa de calor de la tabla de contingencia
plt.figure(figsize=(15, 10))
sns.heatmap(tabla_contingencia_top_10, cmap="YlGnBu", annot=False, fmt='d', linewidths=.5)
plt.title('Correlación entre principales causa de muerte y grupo etareo')
plt.xlabel('Causas de muerte')
plt.ylabel('Grupo de Edad')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

**8. Correlacion entre causas de muerte y enfermedades**

Se observa una alta correlacion de todas las causas de muerte sin distinción de edad salvo la causa de muerte "lesion autoinflingida intencionalmente por ahorcamiento y estrangulamiento" que según el gráfico anterior afectaba principalmente a las edades de 15 a 34 años.

Se excluyen de este analisis dos grupos etareos: de 0 a 14 años y sin especificar.

```python
# Crear un dataframe que excluye la edad sin especificar
edad_sin_edades = ('01.De a 0 a 14 anios', '06.Sin especificar')
df_sin_edades = df_sexo_depurado[df_sexo_depurado['grupo_edad']!=edad_sin_especificar]

# Crear una tabla de contingencia de cie10_clasificacion y grupo_edad
tabla_contingencia = pd.crosstab(df_sin_edades['grupo_edad'], df_sin_edades['cie10_clasificacion'])

# Seleccionar las 10 principales clasificaciones CIE10 para el gráfico
top_10_cie10 = df_sin_edades['cie10_clasificacion'].value_counts().sort_values(ascending=False).head(10).index.tolist()
tabla_contingencia_top_10 = tabla_contingencia[top_10_cie10]

# Calcular la matriz de correlación entre las 10 principales causas de muerte y los grupos de edad
matriz_correlacion = tabla_contingencia_top_10.corr()

# Crear un mapa de calor de la matriz de correlación
plt.figure(figsize=(12, 12))
sns.heatmap(matriz_correlacion, cmap="coolwarm", annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlación entre Top 10 CIE10 y Grupo Etareo')
plt.xlabel('Causas de Muerte (Top 10 CIE10)')
plt.ylabel('Grupo de Edad')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
